Question 1--> What are semantic HTML elements? Why is using them important for web
development?


Answer-->
    Semantic HTML elements are HTML tags that clearly describe their meaning and purpose both to the browser and to developers. In other words, they tell you (and the computer) what kind of content they contain, rather than just how the content looks.

    example of Semantic tag--> Header ,nav ,section ,footer , form , table etc

Question 2: You're designing a blog page. Which semantic elements would you use to
structure the page, and why?

Answer--> When designing a blog page, I would use semantic HTML elements to clearly define the structure and meaning of each part of the page.

    <header> – Used at the top of the page to include the blog title, logo, and main navigation menu.

    <nav> – Contains navigation links (e.g., Home, About, Contact).

    <main> – Holds the main content of the blog page, such as the list of articles or a single blog post.

    <article> – Represents an individual blog post. Each article is self-contained and can stand alone,

    <section> – Used inside an article to divide content into meaningful parts, such as an introduction, body, or conclusion.

    <aside> – Contains related information like recent posts, categories, or advertisements. It’s not part of the main content but provides useful context.

    <figure> and <figcaption> – Used for images, charts, or other visuals with descriptive captions that explain their relevance.

    <footer> – Placed at the bottom of the page or each article to include copyright info, author details, or links to related content.

Question 3: How can you make an HTML form more accessible to users with
disabilities?


Answer-->

Making an HTML form accessible ensures that users with disabilities—such as those using screen readers, keyboard navigation, or other assistive technologies—can easily interact with it. Here are key ways to make a form accessible:


    To make an HTML form accessible:

    Use labels for all inputs

    Include clear instructions and error messages

    Ensure keyboard navigation

    Group fields with <fieldset> and <legend>

    Add ARIA attributes where necessary

    Maintain color contrast and visual clarity

    Never rely on color or placeholder text alone

Question 7: How do the relative, absolute, and fixed positioning properties differ in
CSS?





Answer-->

1. position: relative

The element is positioned relative to its original position in the normal document flow.

It does not remove the element from the document flow.

You can move it using the properties: top, right, bottom, and left.


2. position: absolute

The element is removed from the normal document flow.

It is positioned relative to the nearest positioned ancestor (the closest ancestor with position: relative, absolute, or fixed).

If no such ancestor exists, it is positioned relative to the document body.

3. position: fixed

The element is removed from the document flow.

It is positioned relative to the browser window (viewport), not any parent element.

It stays in the same place even when the page is scrolled.
